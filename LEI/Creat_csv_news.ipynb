
## Criar um ficheiro csv a partir de multiplos JSON
  

```python
import json
import os
from os import listdir
import csv
```

###  Publico

```python
row_list = []
```

```python
row_list.append(["postText;sentimento"])
  
for file in os.listdir(os.getcwd() + "/Publico-Json"):
    print(file)
    
    f = open('Publico-Json/' + file, encoding="utf-8")

    data = json.load(f)
    
    texto = data['header']['postText'].split("['")[1].split("']")[0]
    
    row_list.append(texto)
     
    f.close()
```

```python


print(row_list)


```

```python
with open('publico.csv', 'w', newline='', encoding="utf-8") as file:
    file=csv.writer(file, delimiter=';', quotechar='"', quoting=csv.QUOTE_MINIMAL)

    file.writerows(row_list)
    
```

```python
with open('publico.csv', mode='w',  encoding="utf-8") as csv_file:
    fieldnames = ['texto', 'sentimento']
    
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames,  delimiter=';')

    writer.writeheader()
    for i in row_list[1:]:
        writer.writerow({'texto': i, "sentimento" : 0})
```
